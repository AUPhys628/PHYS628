Figure 1: Magnetic field data from Solar Orbiter, Parker Solar Probe, Magnetic Multiscale Mission, THEMIS, and Arase missions for 2 h on 20 April 2020.

```python
import pyspedas
from pyspedas import tplot, tplot_options
tplot_options('axis_font_size', 10)
time_range = ['2020-04-20/06:00', '2020-04-20/08:00']
pyspedas.projects.solo.mag(trange=time_range, time_clip=True)
pyspedas.projects.psp.fields(trange=time_range, time_clip=True)
pyspedas.projects.mms.fgm(trange=time_range, time_clip=True, probe=2, spdf=True)
pyspedas.projects.themis.fgm(trange=time_range, time_clip=True, probe='d')
pyspedas.projects.erg.mgf(trange=time_range, time_clip=True)
```

```python
tplot(['B_RTN',
'psp_fld_l2_mag_RTN',
'mms2_fgm_b_gsm_srvy_l2_bvec',
'thd_fgs_gsm',
'erg_mgf_l2_mag_8sec_gsm'])
```

FIGURE 3.
From top to bottom, (A) FGM data observed by MMS-2 in GSE coordinates (from the CDF files), (B) MMS-2 FGM data in GSE coordinates transformed from GSM coordinates, (C) Arase MGF data in GSE coordinates (from the CDF files), and (D) Arase MGF data in GSE coordinates transformed from GSM coordinates.


```python
from pyspedas import cotrans
cotrans(name_in='erg_mgf_l2_mag_8sec_gsm',
name_out='erg_mgf_l2_mag_8sec_gsm_2gse',
coord_in='gsm',
coord_out='gse')
cotrans(name_in='mms2_fgm_b_gsm_srvy_l2_bvec',
name_out='mms2_fgm_b_gsm_srvy_l2_bvec_2gse',
coord_out='gse')
```

```python
tplot(['mms2_fgm_b_gse_srvy_l2_bvec',
'mms2_fgm_b_gsm_srvy_l2_bvec_2gse',
'erg_mgf_l2_mag_8sec_gse',
'erg_mgf_l2_mag_8sec_gsm_2gse'])
```
