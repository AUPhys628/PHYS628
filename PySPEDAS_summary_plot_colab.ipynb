### Imports

```python
from pyspedas.projects.themis import fgm, esa, state, efi
from pyspedas import tplot, options, tplot_options, tlimit
```

### Parameters
Edit `trange` and `probe` here to change the event and spacecraft.

```python
trange = ['2014-12-23/11:00:00', '2014-12-23/12:00:00']
probe = 'd'
```

### Load Data
Downloads data from the SPEDAS server. May take a minute or two. Once loaded, you can re-run the plot cell without re-running this cell.

```python
state(probe=probe, trange=trange)
fgm(probe=probe, trange=trange, level='l2')
esa(probe=probe, trange=trange, level='l2')
efi(probe=probe, trange=trange, level='l2')
```

### Plot

```python
tplot_options('title', f'THEMIS-{probe.upper()} Summary Plot')

variables = [
    f'th{probe}_fgs_btotal',        # |B|
    f'th{probe}_fgs_gse',           # Bx, By, Bz
    f'th{probe}_eff_dot0_gse',      # Electric field
    f'th{probe}_peir_en_eflux',     # Ion energy spectrogram
]

tlimit(trange)
tplot(variables)
```
